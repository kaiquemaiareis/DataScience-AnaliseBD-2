```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
# Exibe os nomes das colunas
print(df.columns)
```

```python
# Obtem o numero de entradas, tipos de dados e valores nulos
print(df.info())
# Numero de valores nulos em cada coluna
df.isnull().sum()
```

```python
# Distribuiçao dos valores nas colunas numericas
df.describe()
```

```python
# Configuraçoes gerais dos graficos
plt.figure(figsize=(6, 4))
sns.set(style="whitegrid")

# Distribuiçao da Qualidade do Ar 
plt.subplot(2, 2, 1)
sns.histplot(df['AirQuality'], kde=True, color='blue')
plt.title('Distribuição da Qualidade do Ar')
plt.xlabel('Qualidade do Ar')
plt.ylabel('Frequência')

# Ajusta o layout
plt.tight_layout()
plt.show()
```

```python
# Configuraçoes gerais dos graficos
plt.figure(figsize=(6, 4))
sns.set(style="whitegrid")

# Distribuiçao da Poluiçao da Agua
plt.subplot(2, 2, 2)
sns.histplot(df['WaterPollution'], kde=True, color='red')
plt.title('Distribuição da Poluição da Água')
plt.xlabel('Poluição da Água')
plt.ylabel('Frequência')

# Ajusta o layout
plt.tight_layout()
plt.show()
```

```python
# Correlaçao entre qualidade do ar e poluiçao da agua
correlation = df['AirQuality'].corr(df['WaterPollution'])
print('Correlação entre Qualidade do Ar e Poluição da Água:', correlation)
```

```python
# Configuraçoes gerais dos graficos
plt.figure(figsize=(20, 5))
sns.set(style="whitegrid")

# Filtra as 15 regioes mais comuns
top_regions = df['Region'].value_counts().head(15).index
df_top_regions = df[df['Region'].isin(top_regions)]

# Qualidade do Ar por Regiao (top 15 regioes mais comuns)
plt.subplot(2, 2, 3)
sns.boxplot(x='Region', y='AirQuality', data=df_top_regions)
plt.title('Qualidade do Ar por Região (Top 15)')
plt.xlabel('Região')
plt.ylabel('Qualidade do Ar')
plt.xticks(rotation=45)

# Ajusta o layout
plt.tight_layout()
plt.show()
```

```python
# Configuraçoes gerais dos graficos
plt.figure(figsize=(20, 5))
sns.set(style="whitegrid")

# Filtra as 15 regioes mais comuns
top_regions = df['Region'].value_counts().head(15).index
df_top_regions = df[df['Region'].isin(top_regions)]

# Poluiçao da Agua por Regiao (15 regioes mais comuns)
plt.subplot(2, 2, 4)
sns.boxplot(x='Region', y='WaterPollution', data=df_top_regions)
plt.title('Poluição da Água por Região (Top 15)')
plt.xlabel('Região')
plt.ylabel('Poluição da Água')
plt.xticks(rotation=45)

# Ajusta o layout
plt.tight_layout()
plt.show()
```

```python
# Configuraçoes gerais dos graficos
plt.figure(figsize=(9, 6))
sns.set(style="whitegrid")

# Filtrar os países mais comuns (os 10 países com mais registros)
top_countries = df['Country'].value_counts().head(10).index
df_top_countries = df[df['Country'].isin(top_countries)]

# Comparaçao entre Qualidade do Ar e Poluiçao da Agua por País
sns.scatterplot(x='AirQuality', y='WaterPollution', hue='Country', data=df_top_countries)
plt.title('Comparação entre Qualidade do Ar e Poluição da Água por País (Top 10 Países)')
plt.xlabel('Qualidade do Ar')
plt.ylabel('Poluição da Água')
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Ajusta o layout
plt.tight_layout()
plt.show()
```
